```python
# import the Python packages that we need

import sweetviz as sv
from sklearn import datasets
```

```python
df_housing = datasets.fetch_california_housing(as_frame=True).frame
```

```python
df_housing.head(10)
```

```python
sv_report = sv.analyze(df_housing)
```

```python
sv_report.show_html(  
            filepath='SWEETVIZ_REPORT_1.html', 
            open_browser=True, 
            layout='widescreen', 
            scale=None)
```

```python
# create two random samples

df_housing_sample1 = df_housing.sample(frac=0.5, random_state = 123 )
df_housing_sample2 = df_housing.drop(df_housing_sample1.index)
```

```python
print(len(df_housing_sample1))
print(len(df_housing_sample2))
print(len(df_housing))
```

```python
# compare the two samples with each other

sv_report_comparison = sv.compare(df_housing_sample1, df_housing_sample2)
sv_report_comparison.show_html('SWEETVIZ_REPORT_COMPARISON.html')
```
